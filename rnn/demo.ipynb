```python
import tensorflow as tf
import rnn as rnn

def basic_rnn_demo():
    sess, x, y ,pred= rnn.train(training_iters = 1000)
    rnn.prediction(sess, x, y, pred )
    sess.close()

def lstm_rnn_demo():
    sess, x, y ,pred= rnn.train(training_iters = 1000,use_lstm=True)
    rnn.prediction(sess, x, y, pred )
    sess.close()
```

```python
with tf.variable_scope("basic"):
    basic_rnn_demo()
    
```

```python
with tf.variable_scope("lstm"):
    lstm_rnn_demo()
```
