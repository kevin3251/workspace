```python
%matplotlib inline
import numpy as np
import mlp as mlp
import matplotlib.pyplot as plt
```

```python
i,o = 16,10
nl = np.array([i,100,100,o])
v1 = np.zeros(20)
for e in range(20):
    print("epoch:", e+1)
    mlp.train(nl, e+1, 1)
    v1[e] = mlp.predict(nl)
print(v1)
plt.plot(v1)
plt.show()
```
